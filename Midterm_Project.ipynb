## Data Cleaning and Wrangling

I import my data from github, and read my data

```python
import os as path
pd.set_option('display.max_columns',None)
path=('/content/drive/MyDrive/Dataset/Crime_Data_from_2020_to_Present.csv')
crime_df=pd.read_csv(path)
crime_df
```

I inspect my dataset data types

```python
crime_df.dtypes
```

I change the data types to string


```python
crime_df = crime_df.astype({
    "Date Rptd": "string",
    "DATE OCC": "string",
    "AREA NAME" : "string",
    "Crm Cd Desc" : "string",
    "Mocodes" : "string",
    "Vict Sex" : "string",
    "Vict Descent" : "string",
    "Premis Desc" : "string",
    "Weapon Desc" : "string",
    "Status" : "string",
    "Status Desc" : "string",
    "LOCATION" : "string",
    "Cross Street" : "string",
    "Crm Cd 3" : "string",
    "Crm Cd 4" : "string"
})
crime_df.dtypes
```

I identify any missing data

```python
#crime_df.isnull().sum()
crime_df.isnull().sum().sort_values(ascending=False) / crime_df.shape[0]
```

```python
#crime_df['Vict Descent'] = crime_df['Vict Descent'].astype(str).str.replace('O', np.nan)
crime_df['Vict Descent'] = crime_df['Vict Descent'].astype(str).str.replace('X','nan')
crime_df['Vict Sex'] = crime_df['Vict Sex'].astype(str).str.replace('X', 'nan')
crime_df['Vict Age'] = crime_df['Vict Age'].astype(str).str.replace('7Un', '70')

crime_df.replace("N/A", np.nan, inplace=True) #replaces every "N/A" with nan so that we can filter
crime_df.replace("nan", np.nan, inplace=True) #replaces every "nan" with nan so that we can filter
crime_df['Vict Age'] = pd.to_numeric(crime_df['Vict Age'], errors='coerce') #removes any value in Vict Age that had NaN attached and converts to numeric

crime_df
```

Removing missing data (rows)

```python
crime_df.dropna(inplace = True)
crime_df
```

```python
crime_df.drop(columns = ["Date Rptd"])
```

# Measure of Central Tendency

I calculated the mean, median and mod of the Premis Cd column

```python
mean_Premis_Cd = crime_df['Premis Cd'].mean()
median_Premis_Cd = crime_df['Premis Cd'].median()
mode_Premis_Cd = crime_df['Premis Cd'].mode().iloc[0]
mean_Premis_Cd,median_Premis_Cd, mode_Premis_Cd
```

### Measures of Spread


I calculated the standard deviation, variance, Range and IQR for the Premis Cd column

```python
Premis_Cd_range = crime_df['Premis Cd'].max() - crime_df['Premis Cd'].min()
Premis_Cd_sample = crime_df['Premis Cd'].var()
std_sample = crime_df['Premis Cd'].std()
Q1 = crime_df['Premis Cd'].quantile(0.25)
Q3 = crime_df['Premis Cd'].quantile(0.75)

IQR = Q3 - Q1
Premis_Cd_range,Premis_Cd_sample,std_sample,IQR

```

Scatterplot with Linear Regression

```python
plt.scatter(crime_df["AREA"],
            crime_df["Rpt Dist No"],
            color='blue' ,
            alpha=0.7,
            edgecolors='black')
plt.plot = np.polyfit(crime_df["AREA"],
                      crime_df["Rpt Dist No"],
                      deg=1)
x = crime_df['AREA'].to_numpy().reshape(-1, 1)
y = crime_df['Rpt Dist No'].to_numpy()
lin_reg = LinearRegression()
lin_reg.fit(x,y)
beta_0 = lin_reg.intercept_
beta_1 = lin_reg.coef_[0]
print(f"Intercept (Î²â‚€): {beta_0:.2f}")
print(f"Slope (Î²â‚): {beta_1:.2f}")
plt.figure(figsize=(10, 6))
sns.regplot(x=x[:, 0], y=y, scatter=True, ci=None, line_kws={"color": "red"})

# Create the equation text and annotate the plot
equation_text = f"$y = {beta_0:.2f} + {beta_1:.2f}x$"
plt.text(0.05, 0.95, equation_text, transform=plt.gca().transAxes, fontsize=12, color="red", verticalalignment='top')
plt.title('Linear Regression Fit')
plt.xlabel('AREA')
plt.ylabel('Rpt Dist No')
plt.show()
```

```python
crime_df_Vict = crime_df[(crime_df['Vict Sex'] == 'F') & (crime_df['Crm Cd Desc'] =='RAPE, FORCIBLE')]

# Calculate the probability
prob_Female_crime_Vict = len(crime_df_Vict) / len(crime_df[crime_df['Crm Cd Desc'] == 'RAPE, FORCIBLE'])

# Print the probability
print(f"The probability of a person being Female who has suffered harm because of a crime committed to them: {prob_Female_crime_Vict*100:.2f}%")
```

```python
hispanic_victims = len(crime_df[crime_df['Vict Descent'] == 'H'])
total_crimes = len(crime_df)
probability = hispanic_victims / total_crimes
print(f"The probability is: {probability:.2f}")
```

```python
battery_area2_crimes = len(crime_df[(crime_df['Crm Cd Desc'] == 'BATTERY') & (crime_df['AREA'] == 2)])
total_crimes = len(crime_df)
probability = battery_area2_crimes / total_crimes
print(f"The probability is: {probability:.2f}")
```

```python
residence_female_crimes = len(crime_df[(crime_df['Premis Desc'] == 'RESIDENCE') & (crime_df['Vict Sex'] == 'F')])
total_crimes = len(crime_df)
probability = residence_female_crimes / total_crimes
print(f"The probability is: {probability:.2f}")
```
